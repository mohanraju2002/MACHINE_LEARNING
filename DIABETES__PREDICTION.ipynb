# **DIABETES PREDICTION BY USING MACHIONE LEARNING**

**STEP 1: IMPORT THE LIBRARIES**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

**STEP 2:LOAD THE DATA**

```python
from google.colab import files
upload=files.upload()
```

```python
# UNDERSTANDING THE DATA_FRAME

df=pd.read_csv("diabetes_dataset_600_rows.csv")
print("shape:",df.shape)
print("columns:",df.columns)
```

```python
# PRINT THE TOP FIVE RECORDS
df.head()
```

```python
# GETTING THE STATISTICAL MEASURE OF THE DATA
df.describe()
```

```python
# TO FIND THE NUMBER OF PEOPLES SUFFERED FROM DIABETES
df["Outcome"].value_counts()
```

1 --> DIABETES

O --> NO DIABATES

```python
# GROUP BY  DIABETES AND NO DIABATES
df.groupby("Outcome").mean()
```

**STEP 3:DATA PREPARATION**

```python
X=df.drop("Outcome",axis="columns")
y=df["Outcome"]
```

```python
X
```

```python
y
```

```python
# STANDARIZED THE DATA
scalar=StandardScaler()
standarized_data=scalar.fit_transform(X)
print(standarized_data)
```

```python
X=standarized_data
y=df["Outcome"]
print(X)
print(y)
```

**STEP 4: SPLITTING THE DATA**

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

**STEP 5:MODEL CREATION**

```python
model=SVC()
```

**STEP 6: TRAIN THE MODEL**

```python
model.fit(x_train,y_train)
a=model.predict(x_train)
print(accuracy_score(a,y_train))
```

**STEP 7: PREDICT THE MODEL**

```python
print(model.predict([[6,118,13,10,25,21,0.139,57]]))
```

# **CHECKING THE ERRORS**

**MEAN SQUARE ERROR**

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(a,y_train)
print(mse)
```

**ROOT MEAN SQUARE ERROR**

```python
from sklearn.metrics import mean_squared_error
import numpy as np
mse=mean_squared_error(a,y_train)
rmse=np.sqrt(mse)
print(rmse)
```

**MEAN ABSOLUTE ERROR**

```python
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(a,y_train)
print(mae)
```
